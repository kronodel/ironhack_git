# Label Encoding

```python
cols = ["TransactionID", "ClientID", "Profession", "Bank_dep", "Risk", "Number of credits", "Revenue"]
row_1 = [1,231,"Self-employed", 9, "High", 2, 30200]
row_2 = [2,765,"students",5,"high",3,12700]
row_3 = [3,453,"Horeca",7,"medium",5,89400]
row_4 = [4,231,"self-employed",9,"high",2,30200]
row_5 = [5,892,"finance",3,"low",3,740000]
data = pd.DataFrame([row_1,row_2,row_3,row_4,row_5], columns=cols)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
lab_encoder = label_encoder.fit_transform(data['Risk'])
```

# Ordinal Encoding

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

cols = ["TransactionID", "ClientID", "Profession", "Bank_dep", "Risk", "Number of credits", "Revenue"]
row_1 = [1,231,"Self-employed", 9, "High", 2, 30200]
row_2 = [2,765,"students",5,"high",3,12700]
row_3 = [3,453,"Horeca",7,"medium",5,89400]
row_4 = [4,231,"self-employed",9,"high",2,30200]
row_5 = [5,892,"finance",3,"low",3,740000]
data = pd.DataFrame([row_1,row_2,row_3,row_4,row_5], columns=cols)

enc = OrdinalEncoder()
encoded_data = enc.fit_transform(data[["Profession","Risk"]].values)
encoded_df = pd.DataFrame(encoded_data, columns=["Profession_encoded","Risk_encoded"])
data = pd.concat([data, encoded_df], axis=1)
data = data.drop(["Profession","Risk"], axis=1)

print(data)
```
