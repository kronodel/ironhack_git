```python
results = response.json()
flattened_data = pd.json_normalize(results)

flattened_data
```

```python
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
nested_columns = ['actor', 'org', 'payload', 'repo']

flat = flatten(data, nested_columns)


flat.head(10)
```

```python

query = 'covid -is:retweet has:media'
tweets = client.search_recent_tweets(query=query, tweet_fields=['context_annotations', 'created_at'],
                                media_fields=['preview_image_url'], expansions='attachments.media_keys', max_results=100)
# Get list of media from the includes object
media = {m["media_key"]: m for m in tweets.includes['media']}
for tweet in tweets.data:
    attachments = tweet.data['attachments']
    media_keys = attachments['media_keys']
    print(tweet.text)
    if media[media_keys[0]].preview_image_url:
         print(media[media_keys[0]].preview_image_url)
```

```python
# Only tweets in english

query = 'covid -is:retweet has:media lang:en' #lang:en says that we only want english tweets
tweets = client.search_recent_tweets(query=query, tweet_fields=['context_annotations', 'created_at'],
                                media_fields=['preview_image_url'], expansions='attachments.media_keys', max_results=100)
media = {m["media_key"]: m for m in tweets.includes['media']}
for tweet in tweets.data:
    attachments = tweet.data['attachments']
    media_keys = attachments['media_keys']
    print(tweet.text)
    if media[media_keys[0]].preview_image_url:
         print(media[media_keys[0]].preview_image_url)
```

```python
import json

# Convertir un objet Python en JSON
person = {
    "name": "John Smith",
    "age": 30,
    "isStudent": False,
    "courses": ["math", "history", "physics"]
}

json_data = json.dumps(person)
print(json_data)
```

```python
# Convertir un JSON en objet Python
json_data = '{"name": "John Smith", "age": 30, "isStudent": false, "courses": ["math", "history", "physics"]}'
person = json.loads(json_data)
print(person)
```
