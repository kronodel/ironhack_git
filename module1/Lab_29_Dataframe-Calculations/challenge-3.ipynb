You are serving as a Data Analyst at the headquarter of an international fashion goods chain store. Your boss today asked you to do two things for her:

**First, identify two groups of customers from the data set.** The first group is **VIP Customers** whose **aggregated expenses** at your global chain stores are **above the 95th percentile** (aka. 0.95 quantile). The second group is **Preferred Customers** whose **aggregated expenses** are **between the 75th and 95th percentile**.

## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

Next, import `Orders` from Ironhack's database into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
# your code here

orders = pd.read_csv('Orders.csv')
orders.head(10)
```

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
customer_spent = orders.groupby(by=['CustomerID','Country']).sum()
customer_spent.head(10)
```

```python
#Sub Problem 2: How to select customers whose aggregated amount_spent is in a given quantile range?

vip = customer_spent[customer_spent.amount_spent >= customer_spent.amount_spent.quantile(0.95)].index
preferred = customer_spent[(customer_spent.amount_spent >= customer_spent.amount_spent.quantile(0.75)) & (customer_spent.amount_spent < customer_spent.amount_spent.quantile(0.95))].index
```

```python
#Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

customer_spent['customer_type'] = 'normal'
customer_spent.loc[vip, 'customer_type'] = 'vip'
customer_spent.loc[preferred, 'customer_type'] = 'preferred'

customer_spent
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here

# group the data by country and customer type, count the number of customers for each group
grouped = customer_spent.groupby(['Country', 'customer_type']).size().reset_index(name='counts')

# filter for the group with the most VIP customers
vip_group = grouped[grouped['customer_type'] == 'vip'].sort_values('counts', ascending=False).head(1)

# extract the country name from the group with the most VIP customers
most_vip_country = vip_group['Country'].values[0]

print(most_vip_country)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# sum the counts for each country for VIP and Preferred customers
vip_preferred_grouped = grouped[grouped['customer_type'].isin(['vip', 'preferred'])].groupby('Country').sum()

# sort the results in descending order
most_vip_preferred_grouped = vip_preferred_grouped.sort_values('counts', ascending=False)

# extract the country name from the group with the most VIP and Preferred customers combined
most_vip_preferred_country = most_vip_preferred_grouped.index[0]

most_vip_preferred_country 
```
