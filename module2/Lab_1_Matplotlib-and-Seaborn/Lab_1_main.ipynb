#### Plot (x, y) and (x, z).

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x,y,color='r',label='y function',linewidth=2, linestyle='-')

ax[1].plot(x,z,color='g',label='z function',linewidth=2, linestyle='--')
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here
# Method 2
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,6))

ax[0].plot(x,y,color='r',label='y function',linewidth=2, linestyle='-')

ax[1].plot(x,z,color='g',label='z function',linewidth=2, linestyle='--')

ax[0].set_title("y = x * 2")

ax[1].set_title("y = x^2")

plt.tight_layout()

plt.legend()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

# Create figure and subplots
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,5))

#First plot
ax[0].plot(x, x**2, label='y = x^2')
ax[0].plot(x, np.exp(x), label='y = e^x')
ax[0].set_title("Normal Scale")
ax[0].legend()

# Second plot
ax[1].plot(x, x**2, label='y = x^2')
ax[1].plot(x, np.exp(x), label='y = e^x')
ax[1].set_title("Logarithmic Scale")
ax[1].legend()

# Set x-axis and y-axis to logarithmic scale for second subplot
ax[1].set_xscale("log")
ax[1].set_yscale("log")
plt.tight_layout()
# Show plot
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
fig, axes = plt.subplots(1,2,figsize = (12,4))
#fig.suptitle('Median Steps by month VS by work or weekeend', fontsize=14)
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
median_month = fitbit.groupby(['Months_encoded'])['Steps'].median().reset_index()
sns.barplot(x = median_month['Months_encoded'],y = median_month['Steps'],ax=axes[0]).set(xlabel ="Months", ylabel = "Median Steps", title ='Median Steps by Months')
plt.tight_layout()
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
median_work = fitbit.groupby(['Work_or_Weekend'])['Steps'].median().reset_index()
sns.barplot(x = median_work['Work_or_Weekend'],y = median_work['Steps'],ax=axes[1]).set(xlabel ="Work or Weekend", ylabel = "Median Steps", title ='Median Steps by Work or Weekend')
plt.tight_layout()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Very Active vs Steps
