### 1. Import the moovies dataset 

# https://www.kaggle.com/carolzhangdc/imdb-5000-movie-dataset

```python
df = pd.read_csv("movie_metadata.csv")
df.info()
```

```python
df.head(10)
```

### 2. Use crosstabs to show the number of moovies per country

```python
pd.crosstab(index=df["country"], columns="count").sort_values("count", ascending=False)
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
df['director_genre'] = df['director_name'] + "_" + df['genres']
pd.crosstab(index=[df["country"], df["director_genre"]], columns="count")
```

### 4. Average duration of filmes grouped by color

```python
pd.crosstab(index=df['color'], columns="duration", values=df['duration'], aggfunc='mean')
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
pd.crosstab(index=df['director_name'], columns="cast_total_facebook_likes", values=df['cast_total_facebook_likes'], aggfunc='mean')
```
